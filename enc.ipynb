```python

from breezypythongui import EasyFrame
from cryptography.fernet import Fernet
# try importing tkinter to add font and style to the work



# this has to be outside so that same key will be used to encrypt and decrpty
# otherwise it will generate an error
key = Fernet.generate_key()
fernet = Fernet(key)

class Memebrship(EasyFrame):
    def __init__(self):
        EasyFrame.__init__(self,width=300, height=300, title="Member Of a society")
        self.InputField = self.addIntegerField(value="", row=0, column=0, width=10)
        self.addLabel(text="1. Register", row=1, column=0)
        self.addLabel(text="2. Login", row=2, column=0)
        self.addLabel(text="3. Exit", row=3, column=0)
        self.addButton(text="select", row= 4, column =1, command = self.processing)
        
    def processing(self):
        number = self.InputField.getNumber()
        if ( number == 1 ):
            
            class register(EasyFrame):
                def __init__(self):
                    
                    EasyFrame.__init__(self,width=500, height=500, title="Registration")
                    
                    self.addLabel(text="UserName", row=0, column=0)
                    self.InputField1 = self.addTextField(text="", row=0, column=1, width=70)
                    
                    self.addLabel(text="Password", row=1, column=0)
                    self.InputField2 = self.addTextField(text="", row=1, column=1, width=70)
                    
                    self.addButton(text="Submit", row= 2, column =1, command = self.takingPassndUser)
                 
                def takingPassndUser(self):
                    specialCharacter = "$%#@!&%*"
                    pasword_list = []
                    username_list = []
                    
                    userName = self.InputField1.getText()
                    passWord = self.InputField2.getText()
                    
                    
                    if len(passWord) < 5 and not(passWord in (specialCharacter)):
#                         create a popup message
                        print("password length must be greater than 5 and must have a special character")
                    elif len(passWord) < 5:
                        print("password must be greater than 5")
                    elif not(any(specialCharacter) in passWord ):
                        print("password must have a special character")
                    else:
                        pasword_list.append(passWord)
                    username_list.append(userName)
                    
                    
                    userNameCovert = "".join(username_list)
                    passwordConvert = "".join(pasword_list)
                    
                    encMessage = fernet.encrypt(passwordConvert.encode())
                    
                    passwordFile = open("myFile.txt", "wb")
                    passwordFile.write(encMessage)
                    
                    usernameFile = open("userName.txt", "w")
                    usernameFile.write(userNameCovert)
                    
#                   create a pop up message that you have successfully registered

                                      
            register().mainloop()
        if ( number == 2 ):
            class Login(EasyFrame):
                def __init__(self):
                    EasyFrame.__init__(self,width=500, height=500, title="Login")
                    
                    self.addLabel(text="UserName", row=0, column=0)
                    self.InputField3 = self.addTextField(text="", row=0, column=1, width=70)
                    
                    self.addLabel(text="Password", row=1, column=0)
                    self.InputField4 = self.addTextField(text="", row=1, column=1, width=70)
                    
                    self.addButton(text="Submit", row= 2, column =1, command = self.checkingInfo)
                 
                def checkingInfo(self):
                    
                    username1 = self.InputField3.getText()
                    password1 = self.InputField4.getText()
                    
                    file1 = open("userName.txt", "r")
                    file2 = open("myFile.txt", "rb")
                    file1 = file1.read()
#                     print(file2)
                    
#                     the problem
                    file2 = file2.read()
#                     the problem

                    
                    file2= fernet.decrypt(file2).decode()
#                     print(file2)
            
                    if not(username1 == file1) and not(password1 == file2):
                        print("password and username wrong")
                        
                    elif not(password1 == file2):
                        print("Password wrong")
                        
                    elif not(username1 == file1):
                        print("Username wrong")
                        
                    else:
                        print("passoword and username is correct")
                        
            Login().mainloop()
        if ( number == 3 ):
            quit()
#              print("exit")
        
Memebrship().mainloop()
```
